
Data Visualization(minh hoạ dữ liệu) là việc mô tả dữ liệu một cách đơn giản nhất dưới dạng các hình ảnh trực quan như bảng, biểu, đồ thị.

Data Visualization là một bước quan trọng trong quá trình tiền xử lý dữ liệu để xây dựng một mô hình học máy hiệu quả.

Data Visualization giúp hiểu rõ hơn về dữ liệu đầu vào bao gồm: sự phân bố dữ liệu, đặc tính và sự tương quan của các feature, trực quan các dữ liệu bị nhiễu, khuyết thiếu. Những sự hiểu biết này giúp ích rất lớn trong quá trình lựa chọn và huấn luyện mô hình.


Data Visualization được sử dụng trong những giai đoạn nào của một bài toán machine learning?
- Giai đoạn tiền xử lý dữ liệu: Trong giai đoạn này, việc mô hình hóa dữ liệu cung cấp các hiểu biết cơ bản về dữ liệu đầu vào mà chúng ta đang có như sự phân bố, sự khuyết thiếu, độ nhiễu, các giá trị ngoại lai.
- Giai đoạn trích chọn đặc trưng: Mô hình hóa dữ liệu ở giai đoạn này thường kết hợp với các thuật toán ranking/selection feature để đưa ra các biểu đồ đánh giá, so sánh, các bảng score cho từng feature từ đó có thể tính toán được sự phù hợp của feature đối với thuật toán, số lượng feature tối ưu.
- Giai đoạn đánh giá mô hình: Đây là giai đoạn mà việc mô hình hóa dữ liệu được sử dụng nhiều nhất giúp thể hiện kết quả của mô hình học máy, so sánh kết quả giữa các mô hình học máy với nhau, kết quả của quá trình parameter tunning. Đặc biệt việc mô hình hóa dữ liệu trong giai đoạn này còn phục vụ cho quá trình xây dựng các tài liệu business, maketing cho sản phẩm.

Python cung cấp nhiều thư viện đồ họa đi kèm rất nhiều tính năng khác nhau như: Matplotlib, Pandas Visualization, Seaborn, ggplot, Plotly. Trong bài này, chúng ta sẽ học cách tạo biểu đồ cơ bản sử dụng Matplotlib.

```python
import pandas as pd
import seaborn as sns
originalHD = pd.read_csv('HeartDisease.csv')
originalHD.head()
```

Để sử dụng Matplotlib, trước hết ta khai báo

```python
import matplotlib.pyplot as plt
```

---
Trong Matplotlib, chúng ta có thể tạo một biểu đồ histogram dạng biểu đồ cột của các đặc trưng bằng cách gọi phương thức 'hist'. Histogram là một tập hợp các cột biểu thị tần suất xuất hiện của một đại lượng cần theo dõi theo giá trị tăng dần của trục hoành.

```python
plt.figure(figsize = (10, 7)) 
x = originalHD['Age']  
plt.hist(x, bins = 20, color = 'green') 
plt.title('Age') 
plt.xlabel("Age") 
plt.ylabel("Count") 
```

Đồ thị trên mô tả khá rõ phân bố của đặc trưng chol trong tập dữ liệu

-----

Sau đây chúng ta thử xem phân bố của đặc trưng 'fbs' trong tập dữ liệu này

```python
plt.figure(figsize = (10, 7)) 
x = originalHD['fbs']  
plt.hist(x, bins = 20, color = 'green') 
plt.title('fbs') 
plt.xlabel("fbs") 
plt.ylabel("Count")
```

---
Ngoài ra, chúng ta cũng có thể vẽ phân bố của các lớp dữ liệu bị bệnh (chấm đỏ)  và không bị bệnh (dấu sao màu xanh) tương ứng với giá trị của hai đặc đặc trưng (mỗi đặc trưng trên một trục toạ độ).

```python
plt.plot(originalHD.Age[originalHD.num==0],originalHD.chol[originalHD.num==0], 'ro', label='GreenDots')
plt.plot(originalHD.Age[originalHD.num==1],originalHD.chol[originalHD.num==1], '*', label='GreenDots')
plt.xlabel("Age") 
plt.ylabel("chol")
```

---
Biểu đồ đường cũng có thể được sử dụng để quan sát sự biến đổi của các đặc trưng

```python
columns = originalHD.columns.drop(['ID', 'Place', 'Sex', 'num'])
x_data = range(0, originalHD.shape[0])
fig, ax = plt.subplots()

for column in columns:
    ax.plot(x_data, originalHD[column])
# set title and legend
ax.set_title('Heart Disease Dataset')
ax.legend()
```
